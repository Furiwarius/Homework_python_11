Дана функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни                                     

2. Найти интервалы, на которых функция возрастает       

3. Найти интервалы, на которых функция убывает          

4. Построить график                                     

5. Вычислить вершину                                    

6. Определить промежутки, на котором f > 0              

7. Определить промежутки, на котором f < 0              


```python
from sympy import *
from sympy.plotting import plot as plt
init_printing()

x = Symbol('x')
fx = -5*x**2 + 10*x - 150
```

4. Построить график 

```python
p = plt(fx, (x,-7,7))
```

5. Вычислить вершину 

```python
print(solve(diff(fx)))
```

6. Определить промежутки, на котором f > 0

```python
print(solve(fx>0))
```

7. Определить промежутки, на котором f < 0

```python
print(solve(fx<0))
# solve(fx<0)
```

1. Определить корни  

```python
list_x = solve(fx)
for i, elem in enumerate(list_x, 1):
    print(f"{i} корень: {elem}")
```

2. Найти интервалы, на которых функция возрастает

```python
solve(diff(fx)>0)
```

3. Найти интервалы, на которых функция убывает 

```python
solve(diff(fx)<0)
```
